## engagement score


```python
df = pd.read_json('dataset.json')
df = df[["id","inputUrl","username","businessCategoryName","verified","followersCount","igtvVideoCount","relatedProfiles","latestIgtvVideos","postsCount","latestPosts"]]
df = df.dropna(subset=['latestPosts'])
df = df.reset_index(drop=True)
df['latestPosts'].iloc[0]
df['avgLikes'] = [0]*len(df)
df['avgComments'] = [0]*len(df)
df['avgViews'] = [0]*len(df)

```

```python
for i in range(1,len(df)):
    data = df['latestPosts'].iloc[i]
    sumi = 0
    for j in range(len(data)):
        sumi+=data[j]['likesCount']
    df['avgLikes'].iloc[i] = sumi/(len(data)+1)

for i in range(1,len(df)):
    data = df['latestPosts'].iloc[i]
    sumi = 0
    for j in range(len(data)):
        sumi+=data[j]['commentsCount']
    df['avgComments'].iloc[i] = sumi/(len(data)+1)

for i in range(1,len(df)):
    data = df['latestIgtvVideos'].iloc[i]
    sumi = 0
    for j in range(len(data)):
        sumi+=data[j]['likesCount']
    df['avgLikes'].iloc[i] += sumi/(len(data)+1)

for i in range(1,len(df)):
    data = df['latestIgtvVideos'].iloc[i]
    sumi = 0
    for j in range(len(data)):
        sumi+=data[j]['commentsCount']
    df['avgComments'].iloc[i] += sumi/(len(data)+1)

for i in range(1,len(df)):
    data = df['latestIgtvVideos'].iloc[i]
    sumi = 0
    for j in range(len(data)):
        sumi+=data[j]['videoViewCount']
    df['avgViews'].iloc[i] += sumi/(len(data)+1)


for i in range(1,len(df)):
    if(df['avgViews'].iloc[i] == 0):
        df['avgViews'].iloc[i] = df['followersCount'].iloc[i]*0.015
```

## Credibility


```python
def calculate_credibility_score(verified: bool, follower_count: int, num_posts: int, focusRow: str) -> float:
    threshold = df[focusRow].describe()["50%"]
    verified_bonus = 50 if verified==1 else 0
    
    follower_score = 10 * math.log10(follower_count + 1)
    posts_score = 5 * math.log10(num_posts + 1)
    
    post_multiplier = min(1, num_posts / threshold) if threshold > 0 else 1
    adjusted_follower_score = follower_score * post_multiplier
    
    credibility_score = verified_bonus + adjusted_follower_score + posts_score
    return credibility_score
# threshold = 850
df['credibilityScore'] = df.apply(lambda row: calculate_credibility_score(row['verified'], row['followersCount'], row['postsCount'], focusRow="postsCount"), axis=1)
```

```python
from transformers import pipeline

# Initialize the pipeline with your model and label mapping
pipe = pipeline(model="delarosajav95/tw-roberta-base-sentiment-FT-v2")
label_mapping = {"LABEL_0": "Negative", "LABEL_1": "Neutral", "LABEL_2": "Positive"}
tokenizer = pipe.tokenizer  # Retrieve tokenizer from the pipeline

# Assuming df is your DataFrame with columns "username" and "latestComments"
# "latestComments" should be a list of text strings for each user.
df['credScore'] = 0  # Initialize the credibility score column

for user in df["username"].unique():
    # Get the list of texts for this user
    texts = df.loc[df["username"] == user, "latestComments"].values[0]
    
    # Truncate texts that exceed the model's maximum token length
    truncated_texts = []
    for text in texts:
        token_ids = tokenizer.encode(text, add_special_tokens=True)
        if len(token_ids) > tokenizer.model_max_length:
            truncated_text = tokenizer.decode(
                tokenizer.encode(text, truncation=True, max_length=tokenizer.model_max_length),
                skip_special_tokens=True
            )
        else:
            truncated_text = text
        truncated_texts.append(truncated_text)
    
    # Dictionary to count dominant sentiment occurrences
    d = {"Positive": 0, "Negative": 0, "Neutral": 0}
    
    # Process each text individually with error handling
    for text in truncated_texts:
        try:
            # Explicitly pass truncation=True to ensure the text is truncated if necessary
            predictions = pipe(text,truncation=True, top_k=None)
        except Exception as e:
            print("Error processing text:", text)
            print(e)
            continue
        
        # Determine the dominant prediction (highest score) for this text
        mxScore = max(pred.get('score', 0) for pred in predictions)
        for pred in predictions:
            label = label_mapping.get(pred['label'], pred['label'])
            score = pred.get('score', 0)
            if score == mxScore:
                d[label] += 1

    # Calculate overall metrics for the user
    total_comments = d["Positive"] + d["Negative"] + d["Neutral"]
    if total_comments == 0:
        credScore = 50  # Default score if no comments are available
    else:
        if d["Positive"] + d["Negative"] > 0:
            net_sentiment = (d["Positive"] - d["Negative"]) / (d["Positive"] + d["Negative"])
        else:
            net_sentiment = 0
        engagement = (d["Positive"] + d["Negative"]) / total_comments
        credScore = 50 + 50 * (0.7 * net_sentiment + 0.3 * engagement)
        credScore = max(0, min(100, credScore))
    
    print("==================================")
    print(f"User: {user}")
    print(f"Sentiment counts: {d}")
    if total_comments > 0:
        print(f"Net Sentiment: {net_sentiment:.2f}, Engagement: {engagement:.2f}")
    print(f"Credibility Score: {credScore:.2f}")
    
    # Update the DataFrame with the computed credibility score
    df.loc[df["username"] == user, "credScore"] = credScore
```
